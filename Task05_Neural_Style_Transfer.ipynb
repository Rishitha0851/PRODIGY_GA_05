```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

# Load image
img = Image.open(f"/content/{uploaded_filename}")

# Apply artistic effect
styled_img = img.filter(ImageFilter.CONTOUR)

# Save result
styled_img.save("styled_output.png")

# Display
plt.imshow(styled_img)
plt.axis('off')
plt.show()
```

To upload an image from your local machine, you can use `google.colab.files.upload()`. This will open a file selection dialog. Make sure to upload the image you want to process.

```python
from google.colab import files

uploaded = files.upload()

# Get the name of the first uploaded file
if uploaded:
  uploaded_filename = list(uploaded.keys())[0]
  print(f"Uploaded file: {uploaded_filename}")
else:
  uploaded_filename = None
  print("No file uploaded.")
```
